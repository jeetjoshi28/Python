The Python standard library is a collection of built-in modules and packages that provide common functionality (file I/O, math, dates, networking, etc.) without installing extra packages.  
It lets you write useful Python programs quickly by reusing tested, ready-to-use code.

```python
import math
print(math.sqrt(16)) # square root of 16
print(math.pow(2, 3)) # 2 raised to the power of 3
```

```python
import random
print(random.randint(1, 10)) # random integer between 1 and 10

print(random.choice(['apple', 'banana', 'cherry'])) # random choice from a list
```

```python
# File and directory operations
import os
print(os.getcwd()) # get current working directory
```

```python
# datetime
from datetime import datetime, timedelta
print(datetime.now()) # current date and time

yesterday = datetime.now() - timedelta(days=1)
print(yesterday) # date and time for yesterday
```

```python
# time
import time
print(time.time()) # current time in seconds since the epoch
time.sleep(2) # sleep for 2 seconds
print("Awake now!")
```

```python
# regular expressions
import re
patter= r'\d+' # pattern to match one or more digits
text = "The price is 100 dollars"
match = re.search(patter, text)
print(match.group()) # print the matched digits
```
